```python
import pickle
from stable_baselines3.sac.sac import SACMaster
```

```python
directory = "./transfer_learning"
filename = f"{directory}/grid_metrics.pkl"
```

```python
def count_all_passed_sections(values_list):
    
    counter = [0] * 9
    for elem in values_list:
        counter[elem] += 1

    return counter
```

```python
with open(filename, 'rb') as f:
        metrics = pickle.load(f)
```

```python
# Use model name as key for the dict
new_dict = {}
for elem in metrics:
    tmp = elem.copy()
    del tmp['model']
    new_dict.update({elem['model'] : tmp})
```

# Sections

```python
from IPython.display import Image
Image(filename=f'{directory}/racetrack_sections.png') 
```

# Racetrack without transfer learning

```python
best_no_transfer = new_dict['racetrack-v0_0_32env.zip']
```

```python
best_no_transfer['section_passed_start_from']
```

```python
best_no_transfer['section_staying_counter']
```

```python
count_all_passed_sections(best_no_transfer['all_passed_sections'])
```

# Best agent on racetrack with transfer learning with only frames

```python
model_name = "racetrack-v0_1_32env_100_0.6"
```

```python
best_transfer_frames = new_dict[model_name]
```

```python
best_transfer_frames['section_passed_start_from']
```

```python
best_transfer_frames['section_staying_counter']
```
